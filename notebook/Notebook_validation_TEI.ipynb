# Validation d'un fichier xml-tei

```python
from lxml import etree
from pathlib import Path


def valider_xml_tei(xml_path, schema_path):
    """
    Valide un fichier XML-TEI à partir d'un fichier schéma.
    Compatible avec les schémas Relax NG (.rng) et XML Schema (.xsd).
    """

    xml_path = Path(xml_path)
    schema_path = Path(schema_path)

    # Charger le XML
    xml_doc = etree.parse(str(xml_path))

    # Charger le schéma
    schema_doc = etree.parse(str(schema_path))

    # Choisir le type de validation selon l'extension du schéma
    if schema_path.suffix.lower() == ".rng":
        schema = etree.RelaxNG(schema_doc)

    elif schema_path.suffix.lower() == ".xsd":
        schema = etree.XMLSchema(schema_doc)

    else:
        raise ValueError("Format de schéma non reconnu. Utilise .rng ou .xsd")

    # Validation
    est_valide = schema.validate(xml_doc)

    if est_valide:
        print("✅ Le fichier XML-TEI est valide.")
    else:
        print("❌ Le fichier XML-TEI n'est pas valide.")
        print("\nErreurs détectées :")
        for erreur in schema.error_log:
            print(f"- Ligne {erreur.line} : {erreur.message}")

    return est_valide
```

```python
valider_xml_tei(
    xml_path="mon_fichier_tei.xml",
    schema_path="tei_all.rng"
)
```
